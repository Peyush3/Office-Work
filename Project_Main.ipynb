```python
val=data.groupby('Date dd/mm/yyyy').size()
year=[]
number_funding=[]

year_number_dict= val.to_dict()
print(year_number_dict)
```

```python
for a,b in year_number_dict.items():
    number_funding.append(b)
    year.append(a)
    print(a,b)
```

```python
plt.plot(year,number_funding)
plt.title('trend of investments over the years')
plt.xlabel('Year')
plt.ylabel('No. of Fundings')
plt.show()
```

```python
data['Amount in USD'] = pd.to_numeric(data['Amount in USD'].str.replace(',', ''), errors='coerce')

val2=data.groupby('Date dd/mm/yyyy')['Amount in USD'].sum()
year2=[]
value_funding=[]

year_value_dict= val2.to_dict()
print(year_value_dict)
```

```python
for a,b in year_value_dict.items():
    value_funding.append(b)
    year2.append(a)
    print(a,b)
```

```python
plt.plot(year2,value_funding)
plt.title('trend of investments over the years')
plt.xlabel('Year')
plt.ylabel('Value of Fundings')
plt.show()
```

```python
df = pd.read_csv('startup_funding.csv')
df = df.dropna(subset=['City  Location'])
def separateCity(city): 
    return city.split('/')[0].strip() 
df['City  Location']=df['City  Location'].apply(separateCity)
df[df['City  Location']=='bangalore'] = 'Bangalore'
df[df['City  Location']=='Delhi'] = 'New Delhi'

df = df.groupby(by='City  Location').size()
df = df.nlargest(10)
df = df.to_dict()
City=[]
no_of_statup=[]
for i , j in df.items():
    City.append(i)
    no_of_statup.append(j)
    print(i,j)

plt.pie(no_of_statup,labels=City,autopct='%.2f')
plt.show()
```
